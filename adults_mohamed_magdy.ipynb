```python
for col in data.columns:
    # find the mode of each column
    mode = data[col].mode()[0]
    # replace NaN values with the mode
    data[col].fillna(mode, inplace=True)
```

```python
data.info()
```

```python
xobj=data.select_dtypes(include=["object"])
xnon=data.select_dtypes(exclude=["object"])
la=LabelEncoder()
for i in range(xobj.shape[1]):
  xobj.iloc[:,i]=la.fit_transform(xobj.iloc[:,i])
df=pd.concat([xnon,xobj],axis=1)
```

```python
df
```

```python
df.corr()
```

```python
x=df.iloc[:,0:14]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn import preprocessing
```

```python
x1 = preprocessing.normalize(x)
df1=pd.DataFrame(x1,columns=['age','fnlwgt','educational-num','capital-gain','capital-loss','hours-per-week','workclass','education','marital-status','occupation','relationship','race','gender','native-country'])
df1
```

```python
#import matplotlib.pyplot as plt
#plt.scatter(x['age'],y)
#plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df1, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier(criterion="entropy", max_depth=20)

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators=100,random_state=42)
# Train the model on training data
rf.fit(x_train,y_train);
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:", accuracy)
```
